```python
import pandas as pd

# Load matches dataset
matches = pd.read_csv("cleaned_matches.csv")

# Clean column names
matches.columns = matches.columns.str.strip().str.lower()


# üèÜ 1. Matches won per season
season_wins = matches.groupby(["season", "winner"]).size().reset_index(name="wins")
season_wins = season_wins.sort_values(["season", "wins"], ascending=[True, False])

print("‚úÖ Matches Won by Teams in Each Season")
display(season_wins.head(20))  # preview first few seasons


# üèÜ 2. Final Winners (Champion of Each Season)
final_matches = matches.sort_values("date").groupby("season").tail(1)

final_winners = final_matches[["season", "winner"]].reset_index(drop=True)

print("\n‚úÖ IPL Final Winners by Season")
display(final_winners)


# üèÜ 3. Total Championships (Titles per Team)
titles_count = final_winners["winner"].value_counts().reset_index()
titles_count.columns = ["team", "titles"]

print("\n‚úÖ Total IPL Titles per Team")
display(titles_count)
```
