```python
import pandas as pd
import matplotlib.pyplot as plt

# Load matches dataset
matches = pd.read_csv("cleaned_matches.csv")

# Clean column names
matches.columns = matches.columns.str.strip().str.lower()


# 🏟️ 1. Most matches at each venue
venue_matches = matches["venue"].value_counts().reset_index()
venue_matches.columns = ["venue", "matches"]

print("✅ Matches Played at Each Venue")
display(venue_matches.head(20))

# Bar Chart - Top 10 venues
plt.figure(figsize=(12,6))
plt.barh(venue_matches["venue"].head(10), venue_matches["matches"].head(10), color="skyblue")
plt.title("🏟️ Top 10 IPL Venues by Matches Hosted")
plt.xlabel("Number of Matches")
plt.gca().invert_yaxis()
plt.show()

# 🏆 2. Winning teams by venue
venue_wins = matches.groupby(["venue", "winner"]).size().reset_index(name="wins")
venue_top_wins = venue_wins.sort_values(["venue", "wins"], ascending=[True, False])

print("\n✅ Team Wins at Each Venue")
display(venue_top_wins.head(20))


# 📊 3. Venue usage across seasons
venue_season = matches.groupby(["season", "venue"]).size().reset_index(name="matches")
venue_season_summary = venue_season.groupby("venue")["matches"].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
venue_season_summary.head(10).plot(kind="bar", color="orange", edgecolor="black")
plt.title("🏟️ Top 10 Venues Across All Seasons")
plt.xlabel("Venue")
plt.ylabel("Total Matches")
plt.xticks(rotation=45, ha="right")
plt.show()
```
